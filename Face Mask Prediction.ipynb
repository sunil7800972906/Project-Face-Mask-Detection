```python
import cv2
from keras.models import load_model
from keras.preprocessing.image import load_img, img_to_array
import numpy as np
model = load_model('C:\\Users\\alok_\\Desktop\\model\\model_lastest.h5')
img_width, img_hieght = 200, 200
```

```python
model.summary()
```

```python
# Load the Cascade face classifier
face_cascade = cv2.CascadeClassifier('C:\\Users\\alok_\\Desktop\\haarcascade\\haarcascade_frontalface_default.xml')
#startt web cam
#cap = cv2.VedeoCapture(0) # for webcam
cap = cv2.VideoCapture(0) 

img_count_full = 0

font = cv2.FONT_HERSHEY_SIMPLEX

org = (1, 1)
class_label=' '
fontScale = 1
color = (255, 0, 0)
thickness = 2

while True:
    img_count_full += 1
    
    #read image from webcam
    responce, color_img = cap.read()
    
    if responce == False:
        break
        
        
    scale = 100
    width = int(color_img.shape[1] * scale / 100)
    height = int(color_img.shape[0] * scale / 100)
    dim = (width, height)
    
    color_img = cv2.resize(color_img, dim, interpolation=cv2.INTER_AREA)
    
    gray_img = cv2.cvtColor(color_img, cv2.COLOR_BGR2GRAY)
    
    faces = face_cascade.detectMultiScale(gray_img, 1.1, 6)
    
    # take face predict class mask or not mask then draw rectangle and text then display
    img_count = 0
    for (x, y, w, h) in faces:
        org = (x-10,y-10)
        img_count +=1
        color_face = color_img[y:y+h,x:x+w]
        cv2.imwrite('C:\\Users\\alok_\\Desktop\\faces\\input\\%d%dimage.jpg'%(img_count_full,img_count),color_face)
        img = load_img('C:\\Users\\alok_\\Desktop\\faces\\input\\%d%dimage.jpg'%(img_count_full,img_count), target_size=(img_width,img_hieght))
    
        img = img_to_array(img)/255
        img = np.expand_dims(img,axis=0)
        pred_prob = model.predict(img)
        pred=np.argmax(pred_prob)
        
        if pred == 0:
            print("User with mask - predic = ", pred_prob[0][0])
            class_label = "you have wear a mask" 
            speech = gTTS(class_label)
            speech.save("hello.mp3")
            color = (255, 0, 0)
            cv2.imwrite('C:\\Users\\alok_\\Desktop\\faces\\with_mask\\%d%dimage.jpg'%(img_count_full,img_count),color_face)
        else:
            print('user not wearing mask - prob = ', pred_prob[0][1])
            class_label = "Please wear a mask!"
            speech = gTTS(class_label)
            speech.save("hello1.mp3")
            color = (0, 255, 0)
            cv2.imwrite('C:\\Users\\alok_\\Desktop\\faces\\without_mask\\%d%dimage.jpg'%(img_count_full,img_count),color_face)
        cv2.rectangle(color_img, (x,y), (x+w, y+h), (0,0,255), 3)
        # Using cv2.putText() method
        cv2.putText(color_img, class_label, org, font,
                   fontScale, color, thickness, cv2.LINE_AA)
        #display image
        cv2.imshow('LIVE face mask  dectection', color_img)
        
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
            
# Release the VideoCapture object

cap.release()
cv2.destroyAllWindows()    
```
